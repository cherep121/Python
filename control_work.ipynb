№2 Дан одномерный массив числовых значений, насчитывающий N элементов.
Выполнить перемещение элементов массива по кругу вправо, т. е. A[1] → A[2]; A[2] → A[3];
... A[n] → A[1]. 

```python
numbers = list(map(int, input().split()))

print("Исходный массив:", numbers)

if len(numbers) > 0:
    last = numbers[-1]
    for i in range(len(numbers) - 1, 0, -1):
        numbers[i] = numbers[i - 1]
    numbers[0] = last

print("Массив после сдвига вправо:", numbers)
```

№5 Напишите функцию, принимающую от 1 до 3 параметров — целых чисел (как
стандартная функция range). Функция должна выдавать список кубов чисел. Если подан 1
аргумент — от 0 до этого числа (не включительно). Если поданы 2 аргумента, то первый
интерпретируется как начальное число, второй — как конечное (не включительно). Если
поданы 3 аргумента, то третий аргумент интерпретируется как шаг. 

```python
def cube_range(*args):
    if len(args) == 1:
        start, stop, step = 0, args[0], 1
    elif len(args) == 2:
        start, stop, step = args[0], args[1], 1
    elif len(args) == 3:
        start, stop, step = args[0], args[1], args[2]
    else:
        return []
    
    result = []
    i = start
    while i < stop:
        result.append(i ** 3)
        i += step
    return result

print(cube_range(5))
print(cube_range(2, 5))
print(cube_range(1, 10, 2)) 
```

№6 Дана строчка текста, состоящая из слов и пробелов. Словом считается
последовательность символов, слова разделены одним или большим числом пробелов.
Для каждого слова этого текста узнайте порядковый номер его вхождения в текст именно в
той форме, в которой указано слово. Для первого вхождения слова выведите «1», для
второго вхождения того же слова выведите «2» и так далее. 

```python
text = input()
print("Введеный текст: ", text)
words = text.split()
word_count = {}

for word in words:
    if word not in word_count:
        word_count[word] = 1
    else:
        word_count[word] += 1
    print(word_count[word], end=" ")
```

№7 Вывести результат нахождения квадратного корня всех элементов списка, которые
больше 55. Предоставить два решения: 1) при помощи функций map и filter, 2) при помощи
списочного выражения. 


```python
import math

numbers = [10, 64, 49, 100, 36, 81, 55, 60, 48, 121]
print("Исходный список:", numbers)

result1 = list(map(math.sqrt, filter(lambda x: x > 55, numbers)))
print("Решение 1 (map и filter):", result1)

result2 = [math.sqrt(x) for x in numbers if x > 55]
print("Решение 2 (списочное выражение):", result2)
```

№8 Напишите функцию count_cond(condition), которая возвращает функцию одного
аргумента n. Эта функция считает, сколько из чисел от 1 до n включительно удовлетворяют
условию condition - предикату от двух аргументов, где первый - это число n, а второй - число
от 1 до n включительно.
Пример: 


```python
def count_cond(condition):
    def counter(n):
        count = 0
        for i in range(1, n + 1):
            if condition(n, i):
                count += 1
        return count
    return counter

count_factors = count_cond(lambda n, i: n % i == 0)
print(count_factors(2))
print(count_factors(4))
print(count_factors(12))
print()
is_prime = lambda n, i: all(i % j != 0 for j in range(2, int(i**0.5) + 1)) and i > 1
count_primes = count_cond(is_prime)
print(count_primes(2))
print(count_primes(3)) 
print(count_primes(4))
print(count_primes(5))
print(count_primes(20))
```

№9 Напишите функцию compose1(f, g), которая возвращает суперпозицию функций
f(g(x)) от x.

```python
def compose1(f, g):
    return lambda x: f(g(x))

add_one = lambda x: x + 1
composed = compose1(add_one, lambda x: x * 2)
print(composed(12))

square = lambda x: x ** 2
multiply_by_3 = lambda x: x * 3
f1 = compose1(square, multiply_by_3)
print(f1(4))  # (4*3)² = 12² = 144

divide_by_2 = lambda x: x / 2
subtract_5 = lambda x: x - 5
f2 = compose1(divide_by_2, subtract_5)
print(f2(15))
```

№10 Реализуйте иерархию классов:
    Автомобиль (базовый класс)
    Легковой
    Грузовой
    Пассажирский
Поля определите самостоятельно. Должен быть доступен расчёт расхода топлива в
зависимости от загрузки и расчёт стоимости и времени поездки. Продемонстрируйте работу
классов. 


```python
class Car:
    def __init__(self, brand, fuel_consumption_per_100km, speed, fuel_price):
        self.brand = brand
        self.fuel_consumption_per_100km = fuel_consumption_per_100km
        self.speed = speed
        self.fuel_price = fuel_price
    
    def fuel_per_km(self):
        return self.fuel_consumption_per_100km / 100
    
    def fuel_cost(self, distance):
        return distance * self.fuel_per_km() * self.fuel_price
    
    def time_cost(self, distance):
        return distance / self.speed
    
    def __str__(self):
        return f"{self.brand} (расход: {self.fuel_consumption_per_100km}л/100км или \
        {self.fuel_per_km():.3f}л/км, скорость: {self.speed}км/ч, топливо: {self.fuel_price} руб/л)"

class PassengerCar(Car):
    def __init__(self, brand, fuel_consumption_per_100km, speed, seats):
        super().__init__(brand, fuel_consumption_per_100km, speed, 61)
        self.seats = seats
    
    def fuel_per_km(self, passengers=1):
        extra = (passengers / self.seats) * 0.2
        return (self.fuel_consumption_per_100km / 100) * (1 + extra)
    
    def fuel_cost(self, distance, passengers=1):
        return distance * self.fuel_per_km(passengers) * self.fuel_price

class Truck(Car):
    def __init__(self, brand, fuel_consumption_per_100km, speed, load_capacity):
        super().__init__(brand, fuel_consumption_per_100km, speed, 78)
        self.load_capacity = load_capacity
    
    def fuel_per_km(self, load=0):
        extra = (load / self.load_capacity) * 0.3
        return (self.fuel_consumption_per_100km / 100) * (1 + extra)
    
    def fuel_cost(self, distance, load=0):
        return distance * self.fuel_per_km(load) * self.fuel_price

class Bus(Car):
    def __init__(self, brand, fuel_consumption_per_100km, speed, capacity):
        super().__init__(brand, fuel_consumption_per_100km, speed, 78)
        self.capacity = capacity
    
    def fuel_per_km(self, passengers=0):
        extra = (passengers / self.capacity) * 0.15
        return (self.fuel_consumption_per_100km / 100) * (1 + extra)
    
    def fuel_cost(self, distance, passengers=0):
        return distance * self.fuel_per_km(passengers) * self.fuel_price

passenger_car = PassengerCar("Chevrolet niva", 8, 120, 5)
truck = Truck("Volvo FH", 25, 90, 20)
bus = Bus("MAN Lion's", 30, 80, 50)

distance = 500


print(f"ПОЕЗДКА НА РАССТОЯНИЕ {distance} км")


print(f"\n{passenger_car}")
print(f"  - без пассажиров: {passenger_car.fuel_cost(distance, passengers=1):.2f} руб")
print(f"    расход на 1 км: {passenger_car.fuel_per_km(1):.3f} л/км")
print(f"  - с 5 пассажирами: {passenger_car.fuel_cost(distance, passengers=5):.2f} руб")
print(f"    расход на 1 км: {passenger_car.fuel_per_km(5):.3f} л/км")
print(f"  - время в пути: {passenger_car.time_cost(distance):.1f} ч")

print(f"\n{truck}")
print(f"  - без груза: {truck.fuel_cost(distance, load=0):.2f} руб")
print(f"    расход на 1 км: {truck.fuel_per_km(0):.3f} л/км")
print(f"  - с грузом 15 тонн: {truck.fuel_cost(distance, load=15):.2f} руб")
print(f"    расход на 1 км: {truck.fuel_per_km(15):.3f} л/км")
print(f"  - время в пути: {truck.time_cost(distance):.1f} ч")

print(f"\n{bus}")
print(f"  - без пассажиров: {bus.fuel_cost(distance, passengers=0):.2f} руб")
print(f"    расход на 1 км: {bus.fuel_per_km(0):.3f} л/км")
print(f"  - с 40 пассажирами: {bus.fuel_cost(distance, passengers=40):.2f} руб")
print(f"    расход на 1 км: {bus.fuel_per_km(40):.3f} л/км")
print(f"  - время в пути: {bus.time_cost(distance):.1f} ч")

print()
print("Сравнение стоимости топлива на 100 км")

print(f"Легковой (без пассажиров): {passenger_car.fuel_cost(100, passengers=1):.2f} руб (АИ-92, 61 руб/л)")
print(f"Грузовой (без груза): {truck.fuel_cost(100, load=0):.2f} руб (дизель, 78 руб/л)")
print(f"Автобус (без пассажиров): {bus.fuel_cost(100, passengers=0):.2f} руб (дизель, 78 руб/л)")
```

№11 Напишите класс WheelBeing (Колесник). Экземпляр класса при инициализации
принимает аргументы: количество насекомых, вид насекомых, объем колесника.
Класс обеспечивает выполнение методов (wb – экземпляр класса):
    экземпляры класса можно сравнивать: сначала по объему, затем по количеству
        насекомых, затем по виду по алфавиту;
    из экземпляра класса можно вычесть число: wb -= number, при этом меняются
        значения атрибутов:
    из количества насекомых вычитается число, если остается значение не меньше 10,
        иначе просто 10;
        вид не меняется;
        объем – половина от исходного (при нечетном числе – меньшая часть);
    к экземпляру можно прибавить другой экземпляр: wb2 = wb + wb1, возвращается
        новый экземпляр с атрибутами:
        количество насекомых и объемы равны средним значениям, округленным до
        целого вниз;
        вид – меньший по алфавиту;
        исходные экземпляры не меняются;
    экземпляр класса можно разделить на число: res = wb / number, возвращается
        список экземпляров класса длиной, равной числу, с атрибутами:
        вид – исходный с добавлением числа 1, 2 и т. д.
        количество – целая часть от деления исходного количества на число;
        объем – такой же, как у исходного;
    экземпляр класса можно вызвать с аргументом-числом – возвращается
        значение количество * число // объем;
    add_insects() – добавить насекомых – вызывается с аргументом-числом, при
        положительном значении добавляет к исходному количеству, при отрицательном
        – уменьшает на переданное значение;
     __str__ – возвращает строку:
        Wheel being with a volume of <объем> accommodates <количество насекомых>
        insects of the species <вид>.
    __repr__ – возвращает строку:
        WheelBeing(<количество насекомых>, '<вид>', <объем>)

```python
class WheelBeing:
    def __init__(self, insects, species, volume):
        self.insects = insects
        self.species = species
        self.volume = volume
    
    def __lt__(self, other):
        if self.volume != other.volume:
            return self.volume < other.volume
        if self.insects != other.insects:
            return self.insects < other.insects
        return self.species < other.species
    
    def __eq__(self, other):
        return (self.volume == other.volume and 
                self.insects == other.insects and 
                self.species == other.species)
    
    def __gt__(self, other):
        return not (self < other or self == other)
    
    def __le__(self, other):
        return self < other or self == other
    
    def __ge__(self, other):
        return not self < other
    
    def __ne__(self, other):
        return not self == other
    
    def __isub__(self, number):
        self.insects -= number
        if self.insects < 10:
            self.insects = 10
        self.volume = self.volume // 2
        return self
    
    def __add__(self, other):
        new_insects = (self.insects + other.insects) // 2
        new_volume = (self.volume + other.volume) // 2
        new_species = min(self.species, other.species)
        return WheelBeing(new_insects, new_species, new_volume)
    
    def __truediv__(self, number):
        result = []
        part_insects = self.insects // number
        for i in range(number):
            new_species = self.species + str(i + 1)
            result.append(WheelBeing(part_insects, new_species, self.volume))
        return result
    
    def __call__(self, number):
        return self.insects * number // self.volume
    
    def add_insects(self, number):
        self.insects += number
    
    def __str__(self):
        return f"Wheel being with a volume of {self.volume} accommodates {self.insects} insects of the species {self.species}."
    
    def __repr__(self):
        return f"WheelBeing({self.insects}, '{self.species}', {self.volume})"

wb = WheelBeing(120, "spider", 15)
wbl = WheelBeing(120, "fly", 15)
print(wb > wbl, wb != wbl, wb < wbl)
print(wb, wbl, sep="\n")
print()

wb -= 117
wbl.add_insects(-35)
wb2 = wb + wbl
print(wb, wbl, wb2, sep="\n")
print()

wb = WheelBeing(500, "gnat", 8)
wbl = WheelBeing(498, "midge", 3)
print(wb >= wbl, wb == wbl, wb < wbl)
print(wb, wbl, sep="\n")
print()

wb2 = wb + wbl
print(wb2, wb(3), wbl(17), sep="\n")
res = wb / 3
res[0] -= 10
print(res, wb, sep="\n")
```
