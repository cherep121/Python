Напишите функцию генератор factorials(n), генерирующую последовательность факториалов натуральных чисел.

```python
def factorials(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
        yield fact

# Пример использования
n = int(input())
print(n, "! = ", ' '.join(str(x) for x in factorials(n)))
```

Напишите функцию генератор square_fibonacci(n), генерирующую последовательность квадратов чисел Фибоначчи.

```python
def square_fibonacci(n):
    a, b = 1, 1
    for i in range(n):
        yield a ** 2
        a, b = b, a + b

n = int(input())
print(' '.join(str(x) for x in square_fibonacci(n)))
```

Напишите функцию arithmetic_operation(operation), которая принимает символ одной из четырех арифметических операций, а возвращает функцию двух аргументов для соответствующей операции

```python
def arithmetic_operation(operation):
    if operation == '+':
        return lambda x, y: x + y
    elif operation == '-':
        return lambda x, y: x - y
    elif operation == '*':
        return lambda x, y: x * y
    elif operation == '/':
        return lambda x, y: x / y


operation = arithmetic_operation('+')
print("8+3 =", operation(8, 3))
operation = arithmetic_operation('-')
print("12-11 =", operation(12, 11))
operation = arithmetic_operation('*')
print("10*10 =", operation(10, 10))
operation = arithmetic_operation('/')
print("4/2 =", operation(4, 2))
```

Напишите функцию same_by(characteristic, objects), которая проверяет, все ли объекты имеют одинаковое значение некоторой характеристики, и возвращает True, если это так. Если значение характеристики для разных объектов отличается – то False. Для пустого набора объектов, функция должна возвращать True. Аргумент characteristic – это функция, которая принимает объект и вычисляет его характеристику.

```python
def same_by(characteristic, objects):
    if not objects:
        return True
    first_value = characteristic(objects[0])
    return all(characteristic(obj) == first_value for obj in objects)


values1 = [0, 2, 10, 6]
if same_by(lambda x: x % 2, values1):
    print('same')
else:
    print('different')

values2 = [1, 2, 3, 4]
if same_by(lambda x: x % 2, values2):
    print('same')
else:
    print('different')
```

Напишите функцию print_operation_table(operation, num_rows=9, num_columns=9), которая принимает в качестве аргумента функцию, реализующую бинарную операцию вычисления элемента по номеру строки и столбца. Аргументы num_rows и num_columns указывают число строк и столбцов таблицы, которые должны быть распечатаны. Нумерация строк и столбцов идёт с единицы.

Примечание: бинарной операцией называется любая операция, у которой ровно два аргумента, как, например, у операции умножения.

```python
def print_operation_table(operation, num_rows=9, num_columns=9):
    for i in range(1, num_rows + 1):
        row = []
        for j in range(1, num_columns + 1):
            row.append(str(operation(i, j)))
        print(' '.join(row))

print_operation_table(lambda x, y: x * y)
print()
print_operation_table(lambda x, y: x * y, 5)
```

Напишите функцию ask_password(login, password, success, failure), которая получает логин и пароль пользователя и проверяет их правильность. Пароль считается правильным, если в нём содержится ровно три английские гласные буквы (гласными считать буквы a,e,i,o,u,y) и ровно такой же набор согласных (все буквы, кроме перечисленных шести), как в логине. Порядок и количество согласных также должно совпадать. Логин и пароль приводятся к нижнему регистру и передаются.

Пример: для логина “login” подойдут пароли “aaalgn” и “luagon”.

success и failure – коллбэки. Коллбэком называется специальная функция, которая вызывается, когда ваше вычисление завершилось. Для программы, запустившей запрос или долгое вычисление, коллбэк – это способ сообщить, что надлежит сделать, когда вычисление завершится. Коллбэк обычно передают как аргумент функции запуска вычисления.

Если пароль правильный, функция должна вызвать коллбэк success, передав ему в качестве аргумента логин. А если пароль был неверный - failure, передав ему в качестве аргументов логин и сообщение об ошибке (в таком порядке). Сообщение об ошибке должно быть одним из трёх вариантов:

    *  “Wrong number of vowels”, если в пароле неверное число гласных;

    *  “Wrong consonants”, если в пароле набор согласных отличается от набора согласных логина;

    *  “Everything is wrong”, если оба условия нарушены.

Также напишите функцию main(login, password), которая вызывает написанную функцию ask_password так, чтобы в случае успеха она печатала “Привет, {логин}!”, а в случае ошибки – “Кто-то пытался притвориться пользователем {логин}, но в пароле допустил ошибку: {текст ошибки, большими буквами}.”

```python
def ask_password(login, password, success, failure):
    login = login.lower()
    password = password.lower()
    
    vowels = set('aeiouy')
    
    password_vowels = sum(1 for char in password if char in vowels)
    
    login_consonants = ''.join(char for char in login if char not in vowels)
    password_consonants = ''.join(char for char in password if char not in vowels)
    
    vowel_correct = (password_vowels == 3)
    consonant_correct = (login_consonants == password_consonants)
    
    if vowel_correct and consonant_correct:
        success(login)
    elif not vowel_correct and not consonant_correct:
        failure(login, "Everything is wrong")
    elif not vowel_correct:
        failure(login, "Wrong number of vowels")
    else:
        failure(login, "Wrong consonants")

def main(login, password):
    def success(login):
        print(f"Привет, {login}!")
    
    def failure(login, error):
        print(f"Кто-то пытался притвориться пользователем {login}, но в пароле допустил ошибку: {error.upper()}.")
    
    ask_password(login, password, success, failure)

main("anastasia", "nysatos")
main("eugene", "aangi")
ask_password("anastasia", "nysatos", lambda login: print('super'), lambda login, err: print('bad'))
```

Напишите программу для обработки текста. На вход вашей программы подаётся многострочный текст, причём число строк заранее неизвестно.

Ваша задача – пронумеровать слова в нём, начиная с нуля, а затем вывести только те слова, которые начинаются с большой буквы. Перед словом необходимо вывести номер первого вхождения этого слова в текст.

Слова необходимо отсортировать в лексикографическом порядке.

Для решения этой задачи вам помогут итераторы sys.stdin и enumerate.

```python
import sys


words_dict = {}
first_occurrence = {}
    
test_lines = [
    "Ехал Грека через реку.",
    "Видит грека в реке рак.", 
    "Cунул в реку руку Грека",
    "Рак за руку Греку цап"
]
    
for line in test_lines:
    line_words = line.strip().split()
    for word in line_words:
        if word not in words_dict:
            words_dict[word] = 1
            first_occurrence[word] = len(words_dict) - 1
        else:
            words_dict[word] += 1
    
capital_words = [word for word in words_dict.keys() if word and word[0].isupper()]
capital_words.sort(key=lambda x: x.lower())
    
print("\nРезультат (номер первого вхождения и слово):")
for word in capital_words:
    print(first_occurrence[word], word)

```

При решении этой задачи вам понадобится так называемый тернарный оператор (т.е. оператор с тремя аргументами):

<value_if_true> if <condition> else <value_if_false>

Значение тернарного оператора будет value_if_true, если условие condition верно, иначе будет равно value_if_false. Это позволяет выполнять вычисления, зависящие от условия, одной командой.

Например:

    print("чётное" if (x % 2 == 0) else "нечётное")

напечатает, чётное ли число х.

```python
from functools import reduce

lines = ["котик", "кашалот", "тюлень", "кашалот", "нарвал"]
    
if lines:
    result = reduce(lambda x, y: x if x < y else y, lines)
    print(result)
```

```python
def secure_access(worker):
    def protector(*people, **info):
        print("Введите пароль: ")
        code = input()
        if code == "qwerty123":
            return worker(*people, **info)
        else:
            print("Ошибка! Доступ запрещён.")
            return None
    return protector

@secure_access
def hello_user(username):
    print(f"Здравствуй, {username}! Добро пожаловать!")

print("**** Вход ****")
print("Users Антон")
hello_user("Антон")
print("Users Даниил")
hello_user("Данииил")
```

Напишите генератор декораторов check_password, т. е. функцию, которая возвращает декоратор. Генератор декораторов принимает в качестве параметра пароль, и получившийся декоратор должен закрыть функцию этим паролем.

Декоратор будет применяться следующим образом:

@check_password('password')
def make_burger(typeofMeat, withOnion=False, withTomato=True):
...
То есть при определении функции сначала вызывается функция check_password() с аргументом "password", получается декоратор, затем уже этот получившийся декоратор применяется к функции.

```python
def create_password_protector(correct_code):
    def protector_maker(target_func):
        def secured_executor(*inputs, **settings):
            user_input = input("Введите пароль: ")
            if user_input == correct_code:
                return target_func(*inputs, **settings)
            else:
                print("Пароль не подходит! Доступ отклонён.")
                return None
        return secured_executor
    return protector_maker

@create_password_protector("qwe123")
def math_engine(x, y, action):
    if action == "+":
        return x + y
    elif action == "-":
        return x - y
    elif action == "*":
        return x * y
    elif action == "/":
        return x / y if y != 0 else "На ноль делить нельзя!"

print("Введите пароль: ")
result = math_engine(10, 5, "*")
if result is not None:
    print("10*5 = ?")
    print(f"Ответ: {result}")
```

Напишите декоратор cached, который будет кэшировать результат вызова функции. Пример того, как можно будет использовать ваш декоратор:

@cached
def fib(n):
    if n == 1 or n == 2:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)

```python
def cached(func):
    cache = {}
    def wrapper(n):
        if n not in cache:
            cache[n] = func(n)
        return cache[n]
    return wrapper

@cached
def fib(n):
    if n == 1 or n == 2:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)

# Пример использования
print(fib(10))  # Вычисляется и кэшируется
print(fib(5))   # Берется из кэша
print(fib(10))  # Берется из кэша
```
