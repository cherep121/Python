Задача №1 «Количество различных чисел»
Условие:
    Дан список чисел. Определите, сколько в нем встречается различных чисел.
Примечание. Эту задачу на Питоне можно решить в одну строчку.

```python
numbers = list(map(int, input().split()))
print("Исходная строка", numbers)
unique_numbers = set(numbers)
print("Кол-во не повторяющих",len(unique_numbers))
```

Задача №2 «Количество совпадающих чисел»
Условие:  
    Даны два списка чисел. Посчитайте, сколько чисел содержится одновременно как в первом списке, так и во втором.
Примечание. Эту задачу на Питоне можно решить в одну строчку.

```python
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))
print("1 строка: ", list1, ".   2 строка: ", list2)

matching = set(list1) & set(list2)
print("Кол-во совподающих: ", len(matching))
```

Задача №3 «Пересечение множеств»
Условие:
    Даны два списка чисел. Найдите все числа, которые входят как в первый, так и во второй список и выведите их в порядке возрастания.
Примечание. И даже эту задачу на Питоне можно решить в одну строчку.

```python
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))
print("1 строка: ", list1, ".   2 строка: ", list2)
set1 = set(list1)
set2 = set(list2)

intersection = sorted(set(list1) & set(list2))
print("Пересечение строк: ", *intersection)
```

Задача №4 «Встречалось ли число раньше»
Условие:
    Во входной строке записана последовательность чисел через пробел. Для каждого числа выведите слово `YES` (в отдельной строке), если это число ранее встречалось в последовательности или `NO`, если не встречалось.

```python
numbers = list(map(int, input().split()))
seen = set()
print("Исходная строка", numbers)
for num in numbers:
    if num in seen:
        print("YES")
    else:
        print("NO")
        seen.add(num)
```

Задача №5 «Количество слов в тексте»
Условие:
    Дан текст: в первой строке записано число строк, далее идут сами строки. Определите, сколько различных слов содержится в этом тексте.
    Словом считается последовательность непробельных символов идущих подряд, слова разделены одним или большим числом пробелов или символами конца строки.

```python
n = int(input())
words = set()
print("Кол-во строк: ", n)
for i in range(n):
    line = input().split()
    print("Слово: ", line)
    for word in line:
        words.add(word)

print("Кол-во не повторяющих: ", len(words))
```

Задача №6 «Однофамильцы»
Условие:
    Начальник кадровой службы хочет узнать, сколько мужчин-однофамильцев работает в организации. У него есть список фамилий, и на основании этого списка нужно вычислить количество фамилий, которые совпадают с другими.

    В первой строке указывается количество мужчин - сотрудников организации (N). Затем идут N строк с фамилиями этих сотрудников в произвольном порядке.

    Выводится количество однофамильцев в организации.

```python
n = int(input())
surnames = []
print("Кол-во соотрудников: ", n)

while len(surnames) < n:
    surnames.extend(input().split())

counts = {}
for surname in surnames:
    counts[surname] = counts.get(surname, 0) + 1
print("Все соотрудники: ", surnames)

homonyms_count = 0
for count in counts.values():
    if count > 1:
        homonyms_count += count

print("Кол-во соотрудников с одинаковой фамилией: ", homonyms_count)
```

Задача №7 «Номер появления слова»
Условие:
    В единственной строке записан текст. Для каждого слова из данного текста подсчитайте, сколько раз оно встречалось в этом тексте ранее.
    Словом считается последовательность непробельных символов идущих подряд, слова разделены одним или большим числом пробелов или символами конца строки.

```python
text = input().split()

print("Исходное предложение:", *text)

word_counts = {}
for word in text:
    word_counts[word] = word_counts.get(word, 0) + 1

print("Слова которые повторяются:")
has_repeats = False
for word, total in word_counts.items():
    if total > 1:
        print(word, ":", total, "раз(а)")
        has_repeats = True

if not has_repeats:
    print("  Повторяющихся слов нет")
```

Задача №8 «Словарь синонимов»
Условие:
    Вам дан словарь, состоящий из пар слов. Каждое слово является синонимом к парному ему слову. Все слова в словаре различны. Для слова из словаря, записанного в последней строке, определите его синоним.

```python
n = int(input())
synonyms = {}

for i in range(n):
    word1, word2 = input().split()
    synonyms[word1] = word2
    synonyms[word2] = word1

print("Весь словарь синонимов:")
for word, synonym in synonyms.items():
    print(word, "->", synonym)

query = input()

if query in synonyms:
    print("Синоним слова ", query, ":", synonyms[query])
else:
    print("Слово ", query, "не найдено в словаре синонимов")
```

Задача №9 «Выборы в США»
Условие: 
    В первой строке дано количество записей. Далее, каждая запись содержит фамилию кандидата и число голосов, отданных за него в одном из штатов. Подведите итоги выборов: для каждого из участника голосования определите число отданных за него голосов. Участников нужно выводить в алфавитном порядке.

```python
n = int(input())
votes = {}
print("Кол-во изменений", n)

for i in range(n):
    candidate, count = input().split()
    count = int(count)
    print("Введеные кандидаты", candidate, "голосов ->", count)
    votes[candidate] = votes.get(candidate, 0) + count

print("\nРезультаты выборов:")
for candidate in sorted(votes.keys()):
    print("Кандидат:", candidate, "голосов ->", votes[candidate])
```

Задача №10 «Права доступа»
Условие:
Для каждого файла известно, с какими действиями можно к нему обращаться:
    запись W,
    чтение R,
    запуск X.

В первой строке содержится число N — количество файлов. В следующих N строчках содержатся имена файлов и допустимых с ними операций, разделенные пробелами. Далее указано число M — количество запросов к файлам. В последних M строках указан запрос вида Операция Файл.

Для каждого запроса программа должна возвращать `OK`, если операция допустима, или `Access denied`, если операция недопустима.

```python
operation_to_letter = {
    "read": "R",
    "write": "W", 
    "execute": "X"
}

letter_to_operation = {
    "R": "чтение",
    "W": "запись",
    "X": "запуск"
}

n = int(input())
print(n, "файлов загруженно")

file_permissions = {}

for i in range(n):
    data = input().split()
    filename = data[0]
    permissions = set(data[1:])
    file_permissions[filename] = permissions
    
    perm_names = [letter_to_operation[p] for p in permissions]
    print(f"Файл {i+1}: {filename} - доступ: {', '.join(perm_names)} ({', '.join(permissions)})")

m = int(input())
print(m, "запрос(ов)")

for request_num in range(1, m + 1):
    operation, filename = input().split()
    needed_right = operation_to_letter[operation]
    
    print(f"Запрос {request_num}: {operation} -> {filename}")
    
    if filename in file_permissions:
        if needed_right in file_permissions[filename]:
            print(f"  Результат: OK (операция '{operation}' разрешена)")
        else:
            allowed_ops = [op for op, letter in operation_to_letter.items() 
                          if letter in file_permissions[filename]]
            print(f"  Результат: Access denied (операция '{operation}' не разрешена)")
            print(f"  Разрешённые операции для {filename}: {', '.join(allowed_ops)}")
    else:
        print(f"  Результат: Access denied (файл '{filename}' не найден)")
    
    print()
```
